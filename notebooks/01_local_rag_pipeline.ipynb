# 01 — End-to-End Local RAG Pipeline
Backend-agnostic local RAG with advanced retrieval and evaluation hooks.

```python
import sys
from pathlib import Path
sys.path.append(str(Path('..').resolve() / 'src'))
from proferio import AppConfig
from proferio.pipeline import build_basic_rag_pipeline
from proferio.evals import evaluate_single
```

```python
cfg = AppConfig(backend='ollama', model_name='llama3.1:8b', top_k=5)
ask = build_basic_rag_pipeline('../sample_data/docs', cfg)
```

```python
question = 'What governance principles are required for AI outputs?'
result = ask(question)
print(result['answer'])
print('Contexts:', len(result['contexts']))
```

```python
metrics = evaluate_single(
    question=question,
    answer=result['answer'],
    contexts=result['contexts'],
    expected_terms=['citation', 'source', 'governance']
)
metrics
```
