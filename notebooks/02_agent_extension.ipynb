# 02 — Agent Extension
ReAct-style controllable loop with tools and optional human approval callback.

```python
import sys
from pathlib import Path
sys.path.append(str(Path('..').resolve() / 'src'))
from proferio.config import AppConfig
from proferio.pipeline import build_basic_rag_pipeline
from proferio.tools import ToolRegistry, Tool
from proferio.agent import ReActController
```

```python
cfg = AppConfig()
ask = build_basic_rag_pipeline('../sample_data/docs', cfg)
tools = ToolRegistry()
tools.register(Tool('local_rag', 'Query local KB', lambda q: ask(q)['answer']))
tools.register(Tool('calculator', 'Basic demo math', lambda q: str(eval('1+1'))))
agent = ReActController(tools, max_steps=4)
```

```python
def approval(action: str) -> bool:
    return True

state = agent.run('Summarize the required AI safeguards and give evidence.', approval_callback=approval)
state.final_answer
```
